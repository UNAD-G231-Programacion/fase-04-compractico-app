```python
class Customer():
    def __init__(self, customer_id, customer_name) -> None:
        self.customer_id = customer_id
        self.customer_name = customer_name
        
    def __repr__(self):
        # Representa la clase:
        return f"{__class__.__name__}(customer_id={self.customer_id}, customer_name={self.customer_name})"
```
